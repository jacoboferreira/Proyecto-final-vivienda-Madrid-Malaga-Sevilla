## funciones de deteccionde los diferentes datos

```python
def match_property(property, patterns):
    for pat in patterns:
        match_prop = re.search(pat, property)
        if match_prop:
            return True
    return False
```

```python
def get_number(property):
    nums = re.findall(r'\d+', property)
    if len(nums) >= 2:
        return int(nums[0] + nums[1])  # Une dos números como texto, luego convierte
    elif len(nums) == 1:
        return int(nums[0])
    else:
        return 0
```

```python
def check_property(property, patterns):
    for pat in patterns:
        check = re.search(pat, property)
        if check:
            return 1
    return 0
```

## funciones para sacar la información en columnas nuevas

```python
def get_baños(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['baño']):
            return(get_number(prop.lower().strip()))
```

```python
def get_ascensor(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['ascensor']):
            return(check_property(prop.lower().strip(),['con'] ))
```

```python
def get_años(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['construido en']):
            return(get_number(prop.lower().strip()))
```

```python
def get_terraza(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['terraza']):
            value +=1
    return(value)
```

```python
def get_trastero(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['trastero']):
            value +=1
    return(value)
```

```python
def get_orientacion(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['orientación']):
            return(prop.split(' ', maxsplit=1)[1].strip().strip(', '[0]))
```

```python
def get_piso(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['bajo', 'planta', 'interior', 'exterior']):
            return(prop)
```

```python
def get_habitaciones(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['habitaci']):
            try:
                habitaciones = get_number(prop.lower().strip())
            except:
                habitaciones = prop
            return(habitaciones)
```

```python
def get_metros_2(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['m² construidos']):
            try:
                metros = get_number(prop.lower().strip().split(',')[0])
            except:
                metros = prop
            return(metros)
```

```python
def get_condicion(features):
    for prop in features:
        if match_property(prop.lower().strip(), ['segunda mano', 'promocion de obra nueva']):
            return(prop)
```

```python
def get_balcon(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['balcón']):
            value +=1
    return(value)
```

```python
def get_armario_empotrado(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['armarios empotrados']):
            value +=1
    return(value)
```

```python
def get_jardin(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['jardin']):
            value +=1
    return(value)
```

```python
def get_garaje(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['garaje']):
            value +=1
    return(value)
```

```python
def get_calefaccion(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['calefacción']):
            value +=1
    return(value)
```

```python
def get_aire_acon(features):
    value = 0
    for prop in features:
        if match_property(prop.lower().strip(), ['aire acondicionados']):
            value +=1
    return(value)
```

```python
def extraer_consumo(x):
    if isinstance(x, list):
        x = " ".join(x)
    match = re.search(r'Consumo:\s*\\?n?\s*(\d+)', str(x))
    return int(match.group(1)) if match else None
```

```python
def extraer_emisiones(x):
    if isinstance(x, list):
        x = " ".join(x)
    match = re.search(r'Emisiones:\s*\\?n?\s*(\d+)', str(x))
    return int(match.group(1)) if match else None
```

```python
datos_alquiler['m2'] = datos_alquiler.caracteristicas_basicas.apply(get_metros_2)
datos_alquiler['habitaciones'] = datos_alquiler.caracteristicas_basicas.apply(get_habitaciones)
datos_alquiler['baño'] = datos_alquiler.caracteristicas_basicas.apply(get_baños)
datos_alquiler['terraza'] = datos_alquiler.caracteristicas_basicas.apply(get_terraza)
datos_alquiler['estado'] = datos_alquiler.caracteristicas_basicas.apply(get_condicion)
datos_alquiler['armarios'] = datos_alquiler.caracteristicas_basicas.apply(get_armario_empotrado)
datos_alquiler['orientacion'] = datos_alquiler.caracteristicas_basicas.apply(get_orientacion)
datos_alquiler['construido'] = datos_alquiler.caracteristicas_basicas.apply(get_años)
datos_alquiler['calefaccion'] = datos_alquiler.caracteristicas_basicas.apply(get_calefaccion)
datos_alquiler['ascensor'] = datos_alquiler.caracteristicas_basicas.apply(get_ascensor)
datos_alquiler['garaje'] = datos_alquiler.caracteristicas_basicas.apply(get_garaje)
datos_alquiler['trastero'] = datos_alquiler.caracteristicas_basicas.apply(get_trastero)
datos_alquiler['planta'] = datos_alquiler.caracteristicas_basicas.apply(get_piso)
datos_alquiler['balcon'] = datos_alquiler.caracteristicas_basicas.apply(get_balcon)
datos_alquiler['jardin'] = datos_alquiler.caracteristicas_basicas.apply(get_jardin)
datos_alquiler['aire acondicionado'] = datos_alquiler.caracteristicas_extra.apply(get_aire_acon)
datos_alquiler['consumo_kWh'] = datos_alquiler['caracteristicas_extra'].apply(extraer_consumo)
datos_alquiler['emisiones_CO2'] = datos_alquiler['caracteristicas_extra'].apply(extraer_emisiones)
```
