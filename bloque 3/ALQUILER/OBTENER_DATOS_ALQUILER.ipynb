```python
# Iniciar navegador con undetected_chromedriver

options = uc.ChromeOptions()
# Descomenta si no quieres ver el navegador
browser = uc.Chrome(options=options)

# Extraer los ID de los inmuebles con un limite de 1200

x = 1
ids = []
LIMITE_PISOS = 1200

while True:
    if len(ids) >= LIMITE_PISOS:
        print(f"âœ… LÃ­mite de {LIMITE_PISOS} inmuebles alcanzado.")
        break

    url = f'https://www.idealista.com/alquiler-viviendas/villanueva-de-la-canada-madrid/pagina-{x}.htm'
    print(f"ğŸ•¸ï¸ Visitando: {url}")
    browser.get(url)
    time.sleep(5)

    soup = bs(browser.page_source, 'lxml')

    try:
        pagina_actual = int(soup.find('main', {'class': 'listing-items'})
                            .find('div', {'class': 'pagination'})
                            .find('li', {'class': 'selected'}).text)
    except:
        print("ğŸš« Fin de las pÃ¡ginas o error en la paginaciÃ³n.")
        break

    if x != pagina_actual:
        break

    articles = soup.find('main', {'class': 'listing-items'}).find_all('article')
    for article in articles:
        if len(ids) >= LIMITE_PISOS:
            break
        id_muebles = article.get('data-element-id')
        if id_muebles:
            ids.append(id_muebles)
            print(f"ğŸ“Œ ID encontrado: {id_muebles}")
            time.sleep(random.randint(1, 2))

    x += 1

# Guardar los ID en un DataFrame
ids_casas_C = pd.DataFrame(ids, columns=['id'])

print(ids_casas_C.head(10))
print("Total IDs encontrados:", ids_casas_C.shape[0])

# FunciÃ³n para parsear inmuebles
def parsear_inmueble(id_muebles):
    print(f"\nğŸ“„ Parseando inmueble: {id_muebles}")
    url = f'https://www.idealista.com/inmueble/{id_muebles}/'
    browser.get(url)
    time.sleep(2)
    soup = bs(browser.page_source, 'lxml')

    try:
        titulo = soup.find('span', {'class': 'main-info__title-main'}).text.strip()
        localizacion = soup.find('span', {'class': 'main-info__title-minor'}).text.strip().split(',')[0]
        precio = int(soup.find('span', {'class': 'txt-bold'}).text.replace('.', '').strip())

        # CaracterÃ­sticas bÃ¡sicas
        c1 = soup.find('div', {'class': 'details-property-feature-one'})
        c1_features = c1.find('div', {'class': 'details-property_features'}) if c1 else None
        caracteristicas_basicas = [li.text.strip() for li in c1_features.find_all('li')] if c1_features else []

        # CaracterÃ­sticas extra
        c2 = soup.find('div', {'class': 'details-property-feature-two'})
        caracteristicas_extra = []
        if c2:
            for block in c2.find_all('div', {'class': 'details-property_features'}):
                caracteristicas_extra.extend([li.text.strip() for li in block.find_all('li')])

        ubi = soup.find('div', {'id': 'headerMap'})
        ubicacion_des = [zona.text.strip() for zona in ubi.find_all('li')] if ubi else []

        return {
            'id': id_muebles,
            'titulo': titulo,
            'localizaciÃ³n': localizacion,
            'precio': precio,
            'caracteristicas_basicas': caracteristicas_basicas,
            'caracteristicas_extra': caracteristicas_extra,
            'ubicacion': ubicacion_des
        }

    except Exception as e:
        print(f"âŒ Error al parsear {id_muebles}: {e}")
        return None

# Parsear inmuebles
resultados = []

for id_mueble in ids_casas_C['id']:
    resultado = parsear_inmueble(id_mueble)
    if resultado:
        resultados.append(resultado)
    time.sleep(random.randint(1, 2))

# Convertir a DataFrame
df_casas_C = pd.DataFrame(resultados)

# Guardar a CSV final
df_casas_C.to_csv('casa_alquiler_Villanueva_caÃ±ada.csv', index=False, sep=';', encoding='utf-16')

# Cerrar el navegador
browser.quit()

print("âœ… Proceso completo. Datos guardados en casa_alquiler_Villanueva_caÃ±ada.csv")
print(df_casas_C.shape)
```

# Separar la informaciÃ³n de las columnas caracteristicas basicas y extra

```python

# Leer el archivo original
file_path = "datos_unificados_alquiler.csv"
df = pd.read_csv(file_path, encoding='utf-16', sep=';')

# FunciÃ³n para convertir string con formato de lista a lista real
def parse_list(value):
    try:
        return ast.literal_eval(value)
    except Exception:
        return []

df['caracteristicas_basicas'] = df['caracteristicas_basicas'].apply(parse_list)
df['caracteristicas_extra'] = df['caracteristicas_extra'].apply(parse_list)
df['ubicacion'] = df['ubicacion'].apply(parse_list)

# ExtracciÃ³n de informaciÃ³n de caracteristicas_basicas
def extraer_m2_construidos(lista):
    for item in lista:
        if match := re.search(r'(\d+)\s*mÂ² construidos', item):
            return int(match.group(1))
    return None

def extraer_m2_utiles(lista):
    for item in lista:
        if match := re.search(r'(\d+)\s*mÂ² Ãºtiles', item):
            return int(match.group(1))
    return None

def extraer_habitaciones(lista):
    for item in lista:
        if match := re.search(r'(\d+)\s*habitac?', item):
            return int(match.group(1))
    return None

def extraer_banos(lista):
    for item in lista:
        if match := re.search(r'(\d+)\s*baÃ±[oa]s?', item):
            return int(match.group(1))
    return None

# Aplicar funciones
df['m2_construidos'] = df['caracteristicas_basicas'].apply(extraer_m2_construidos)
df['m2_utiles'] = df['caracteristicas_basicas'].apply(extraer_m2_utiles)
df['habitaciones'] = df['caracteristicas_basicas'].apply(extraer_habitaciones)
df['banos'] = df['caracteristicas_basicas'].apply(extraer_banos)

# Extraer comodidades comunes como variables binarias
comodidades = ['aire acondicionado', 'calefacciÃ³n', 'ascensor', 'piscina', 'jardÃ­n', 'amueblado']

for c in comodidades:
    df[c.replace(" ", "_")] = df['caracteristicas_extra'].apply(
        lambda x: any(c in item.lower() for item in x)
    )

# Guardar DataFrame limpio
df.to_csv("alquiler_datos_limpios.csv", index=False, encoding='utf-8-sig')

# Mostrar un resumen
print(df[[
    'precio', 'm2_construidos', 'm2_utiles', 'habitaciones', 'banos',
    'aire_acondicionado', 'calefacciÃ³n', 'ascensor', 'piscina', 'jardÃ­n', 'amueblado'
]].head())
```
