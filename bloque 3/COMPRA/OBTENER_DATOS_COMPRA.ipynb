```python
import pandas as pd
import re
import ast

# 1. Leer el archivo CSV con codificaciÃ³n correcta
file_path = 'datos_unificados_compra.csv'
df = pd.read_csv(file_path, encoding='utf-16', sep=None, engine='python')

# 2. Crear copia de trabajo
df_limpio = df.copy()

# 3. Convertir columnas tipo lista de texto a listas reales
df_limpio['caracteristicas_basicas'] = df_limpio['caracteristicas_basicas'].apply(ast.literal_eval)
df_limpio['caracteristicas_extra'] = df_limpio['caracteristicas_extra'].apply(ast.literal_eval)
df_limpio['ubicacion'] = df_limpio['ubicacion'].apply(ast.literal_eval)

# 4. FunciÃ³n para extraer caracterÃ­sticas bÃ¡sicas
def extraer_info(lista):
    info = {
        'superficie_m2': None,
        'superficie_util_m2': None,
        'habitaciones': None,
        'baÃ±os': None,
        'terraza': False,
        'estado': None,
        'armarios_empotrados': False,
        'orientacion': None,
        'aÃ±o_construccion': None,
        'calefaccion': None
    }
    for item in lista:
        if 'mÂ² construidos' in item:
            match = re.search(r'(\d+)', item)
            if match:
                info['superficie_m2'] = int(match.group(1))
        if 'mÂ² Ãºtiles' in item:
            match = re.search(r'(\d+)', item)
            if match:
                info['superficie_util_m2'] = int(match.group(1))
        elif 'habitaciones' in item:
            match = re.search(r'(\d+)', item)
            if match:
                info['habitaciones'] = int(match.group(1))
        elif 'baÃ±os' in item:
            match = re.search(r'(\d+)', item)
            if match:
                info['baÃ±os'] = int(match.group(1))
        elif 'Terraza' in item:
            info['terraza'] = True
        elif 'Segunda mano' in item or 'A reformar' in item:
            info['estado'] = item
        elif 'Armarios empotrados' in item:
            info['armarios_empotrados'] = True
        elif 'OrientaciÃ³n' in item:
            info['orientacion'] = item.replace('OrientaciÃ³n ', '')
        elif 'Construido en' in item:
            match = re.search(r'(\d{4})', item)
            if match:
                info['aÃ±o_construccion'] = int(match.group(1))
        elif 'CalefacciÃ³n' in item:
            partes = item.split(': ')
            if len(partes) > 1:
                info['calefaccion'] = partes[1]
    return pd.Series(info)

# 5. FunciÃ³n para extraer caracterÃ­sticas extra
def extraer_extra(lista):
    extra = {
        'aire_acondicionado': False,
        'cert_consumo': None,
        'cert_emisiones': None
    }
    for item in lista:
        if 'Aire acondicionado' in item:
            extra['aire_acondicionado'] = True
        elif 'Consumo' in item:
            match = re.search(r'(\d+)\s*kWh', item)
            if match:
                extra['cert_consumo'] = int(match.group(1))
        elif 'Emisiones' in item:
            match = re.search(r'(\d+)', item)
            if match:
                extra['cert_emisiones'] = int(match.group(1))
    return pd.Series(extra)

# 6. FunciÃ³n para extraer ubicaciÃ³n
def extraer_ubicacion(lista):
    ubicacion = {
        'calle': None,
        'barrio': None,
        'distrito': None,
        'ciudad': None
    }
    for item in lista:
        if 'Calle' in item or 'Plaza' in item:
            ubicacion['calle'] = item
        elif 'Barrio' in item:
            ubicacion['barrio'] = item.replace('Barrio ', '')
        elif 'Distrito' in item:
            ubicacion['distrito'] = item.replace('Distrito ', '')
        elif 'Madrid' in item:
            ubicacion['ciudad'] = item
    return pd.Series(ubicacion)

# 7. Aplicar funciones a las columnas correspondientes
caracteristicas_expandidas = df_limpio['caracteristicas_basicas'].apply(extraer_info)
extra_data = df_limpio['caracteristicas_extra'].apply(extraer_extra)
ubicacion_data = df_limpio['ubicacion'].apply(extraer_ubicacion)

# 8. Unir todos los datos al DataFrame original
df_limpio = pd.concat([df_limpio, caracteristicas_expandidas, extra_data, ubicacion_data], axis=1)

# 9. (Opcional) Guardar el DataFrame final como nuevo CSV
df_limpio.to_csv("datos_limpios_compra.csv", index=False)
```
