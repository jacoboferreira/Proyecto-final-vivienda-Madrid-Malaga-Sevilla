```python
import pandas as pd
import os
import glob
```

```python
# RSeleccionar la carpeta de los archivos
carpeta = "C:\\Users\\LENOVO\\Desktop\\Jacobo\\DATOS PISOS\\MALAGA\\COMPRA" 
os.chdir(carpeta)

# Utilizar todos los archivos.xlsx en la carpeta
archivos_excel = glob.glob("*.xlsx")

# Crear lista para DataFrames
dataframes_procesados = []

# Bucle for
for archivo in archivos_excel:
    print(f"Procesando archivo: {archivo}")
    
    # Leer el archivo Excel
    df = pd.read_excel(archivo, dtype=str)
    df = df.fillna("")

    # Eliminar celdas que contienen "https"
    def remove_https(row):
        return ["" if "https" in str(x) else x for x in row]

    df_cleaned = pd.DataFrame(df.apply(remove_https, axis=1).to_list(), columns=df.columns)

    # Mover datos a la izquierda
    def shift_row_left(row):
        values = [x for x in row if x.strip() != ""]
        empty_count = len(row) - len(values)
        return values + [""] * empty_count

    df_final = pd.DataFrame(df_cleaned.apply(shift_row_left, axis=1).to_list(), columns=df.columns)

    # Filtrar filas
    df_final = df_final[~df_final.iloc[:, 0].str.contains("Destacado", case=False, na=False)]
    df_final = df_final[~df_final.iloc[:, 0].str.contains("Top", case=False, na=False)]
    df_final = df_final[~df_final.iloc[:, 0].str.contains("Top+", case=False, na=False)]

    # Eliminar columnas vacías
    df_final = df_final.loc[:, df_final.iloc[0].str.strip() != ""]

    # Añadir columna con el nombre del archivo
    df_final.insert(0, "Archivo_origen", archivo)

    # Agregar al listado general
    dataframes_procesados.append(df_final)

# Unir todos los DataFrames procesados
df_consolidado = pd.concat(dataframes_procesados, ignore_index=True)

# Guardar en un único archivo Excel
df_consolidado.to_excel("DATOS-MALAGA-P.xlsx", index=False)

print("Proceso completado. Archivo consolidado guardado como 'todo_junto.xlsx'.")
```
