```python
import pandas as pd
import os
import glob
```

```python
# Ruta a la carpeta donde están los archivos Excel
carpeta = "C:\\Users\\LENOVO\\Desktop\\Jacobo\\DATOS PISOS\\MALAGA\\EVOLUTIVO"  # <-- Reemplaza esto con la ruta real
os.chdir(carpeta)

# Obtener todos los archivos .xlsx en la carpeta
archivos_excel = glob.glob("*.xlsx")

# Lista para almacenar todos los DataFrames procesados
dataframes_procesados = []

# Procesar cada archivo
for archivo in archivos_excel:
    print(f"Procesando archivo: {archivo}")
    
    # Leer el archivo Excel
    df = pd.read_excel(archivo, dtype=str)

    # Añadir columna con el nombre del archivo para identificar
    df.insert(0, "Archivo_origen", archivo)

    # Agregar al listado general
    dataframes_procesados.append(df)

# Unir todos los DataFrames procesados
df_consolidado = pd.concat(dataframes_procesados, ignore_index=True)

# Guardar en un único archivo Excel
df_consolidado.to_excel("DATOS-MALAGA-EVOLUTIVO.xlsx", index=False)

print("Proceso completado. Archivo consolidado guardado como 'todo_junto.xlsx'.")
```
